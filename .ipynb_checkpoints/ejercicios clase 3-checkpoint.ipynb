### Ejercicio 1: Fusión de imágenes

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img1 = cv.imread("imgs/wanda-y-vision.jpg")
img2 = cv.imread("imgs/marvel.png")
# cambio de espacio de color
img1 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
img2 = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
# igualar el tamano
fil,cols,chan = img1.shape
img2 = cv.resize(img2,(cols,fil))
```

```python
#Mostramos las imagenes
plt.figure(1)
plt.imshow(img1)
plt.title("Wanda y Vision")
plt.figure(2)
plt.imshow(img2)
plt.title("marvel")
```

```python
#combinar imagenes
img_out = cv.addWeighted(img1,0.8,img2,0.3,0)
plt.imshow(img_out)
```

### Ejercicio 3: Operaciones Bit a bit

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
# cargar nuestras imagenes
img1 = cv.imread("imgs/avengers.jpg")
img2 = cv.imread('imgs/marvel.png')
# cambiando espacio de color
avengers = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
marvel = cv.cvtColor(img2, cv.COLOR_BGR2RGB)
# mostrado las imagenes leidas
plt.figure(1)
plt.imshow(avengers)
plt.figure(2)
plt.imshow(marvel)
```

```python
fil2,col2,_= avengers.shape
fil,col,_ = marvel.shape
marvel = cv.resize(marvel,(col2//2,fil//2))
plt.imshow(marvel)
```

```python
fil, col,_ = marvel.shape
roi = img1[0:fil,0:col]
plt.imshow(roi)
```

```python
# creamos una mascara del logotipo
marvel_gray = cv.cvtColor(marvel, cv.COLOR_RGB2GRAY)
ret,mask = cv.threshold(marvel_gray, 150,255, cv.THRESH_BINARY)
plt.imshow(mask, cmap="gray")
```

```python
# Creamos una mascara invertida
mask_inv = cv.bitwise_not(mask)
plt.imshow(mask_inv, cmap="gray")
```

```python
# tomamos el roi menos la mascara
img1_bg = cv.bitwise_and(roi,roi,mask = mask)
plt.imshow(img1_bg)
```

```python
#tomamos region de intere del logotipo opencv
img2_bg = cv.bitwise_and(marvel,marvel,mask=mask_inv)
plt.imshow(img2_bg)
```

```python
img = img1_bg + img2_bg
plt.imshow(img)
```

```python
avengers[0:fil,0:col]=img
plt.imshow(avengers)
```
