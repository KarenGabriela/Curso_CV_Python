 Funciones para Dibujar Geometrias en OpenCV


Dibujar Lnea

```python
# la funcion cv.line()
# cv.line(img, cordendas iniciales, cordenadas finales,color , grosor )
# importamos librerias
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# crear una imagen
img = np.zeros((520,640,3), np.uint8)
# dibujamos la linea en la imagen
cv.line(img,(50,50),(450,450),(255,0,0),6)
# mostrar la imagen
plt.imshow(img)
```

Dibujar crculo

```python
# la funcion cv.circle()
# Parametros: img, condenadas centrales, radio, color, grosor)
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# crear una imagen
img = np.zeros((520,640,3), np.uint8)
# dibujamos circulo
cv.circle(img,(300,250), 100,(255,0,0),-1)
#mostrar nuestra imagen
plt.imshow(img)
```

Dibujar un Rectangulo


```python
# la funcion cv.rectangle()
# Parametros: img, cordenada esquina superir izquierda, cordenada de la esquina,inferior derecha, color, grosor)
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
# crear una imagen
img = np.zeros((520,640,3), np.uint8)
# dibujamos circulo
cv.rectangle(img,(100,100), (450,450),(0,255,0),5)
#mostrar nuestra imagen
plt.imshow(img)
```

 Captura de Eventos desde el Mouse

```python
import cv2 as cv
clicked = False
# funcion encargada de capturar el click del mouse
def onMouse(evento,x,y,flags,param):
# variable global
global clicked
if(evento ==cv.EVENT_LBUTTONUP):
clicked =True
# creamos objeto cv.VideoCapture()
camCap = cv.VideoCapture(0)
cv.namedWindow("MyWindow")
# configuramos se envie a la funcion onMouse los eventos del mouse
cv.setMouseCallback("MyWindow",onMouse)
print("Mostrando estado de la Camara, Presione en la Ventana para detener")
ret,frame = camCap.read()
while(ret and cv.waitKey(1)==-1 and not clicked):
cv.imshow("MyWindow", frame)
ret,frame = camCap.read()
camCap.release()
cv.destroyAllWindows()
```
